```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

def calculate_R0S(ga2, bt2):
    # Fixed Parameters 
    mu = 0.03
    alpha = 0.15
    ga1 = 0.15
    rho1 = 0.5
    rho2 = 0.5
    bt1 = 3.2
    sg1 = 0.3
    sg2 = 0.46
    
    # Calculation
    term_common_denom = (mu + alpha + bt1 * np.sqrt(np.exp(sg1**2/rho1) - 2*np.exp(sg1**2/4/rho1) + 1))
    term1 = bt1 * mu * np.exp(sg1**2/8/rho1) / ((ga1 + mu) * term_common_denom)
    term2 = alpha * bt2 * mu * np.exp(sg2**2/12/rho2) / ((ga1 + mu) * (ga2 + mu) * term_common_denom)
    
    return term1 + term2

def run_experiment_2():
    print("Running Experiment 2: R0s Sensitivity...")
    
    beta2_vals = np.arange(0.02, 3.1, 0.1)
    gamma2_list = [0.2, 0.1, 0.05, 0.02]
    
    
    plt.figure(figsize=(8, 6), dpi=300)
    
    
    colors = ["#2E6C8F", "#4E9F88", "#D48D3B", "#C95D4F"]
    labels = [r'$\gamma_2 = 0.2$', r'$\gamma_2 = 0.1$', r'$\gamma_2 = 0.05$', r'$\gamma_2 = 0.02$']
    
    for i, ga2 in enumerate(gamma2_list):
        r0_vals = [calculate_R0S(ga2, b) for b in beta2_vals]
        
        plt.plot(beta2_vals, r0_vals, label=labels[i], color=colors[i], 
                 linewidth=2.5, marker='o', markersize=6, alpha=0.8)
        
    
    plt.axhline(y=1.0, color='black', linestyle='--', linewidth=1.5)
    
    plt.text(max(beta2_vals), 1.05, 'Threshold = 1', ha='right', va='bottom', fontsize=12, style='italic')
    
    # Styling
    plt.xlabel(r'$\bar{\beta}_2$', fontsize=14, fontweight='bold')
    plt.ylabel(r'$R_0^S$', fontsize=14, fontweight='bold')
    
    
    
    plt.legend(fontsize=12, frameon=False, loc='upper left')
    
    
    plt.grid(True, linestyle=':', alpha=0.6)
    
    
    plt.tight_layout()
    plt.savefig('R0_Sensitivity.pdf', format='pdf', bbox_inches='tight') 
    plt.show()

if __name__ == "__main__":
    run_experiment_2()
```
